```python
### 构造x刻度字符串列表
```

```python
plt.figure(figsize=(20,10), dpi=100)
x = range(60)
x_label = ["11点{}分".format(i) for i in x[::5]]
y = [random.uniform(15,18) for _ in x]
plt.plot(x,y)
plt.title("what is this?")
plt.yticks(range(0, 40, 5))
plt.xticks(x[::5], x_label)
plt.grid(True, linestyle="--", alpha=0.8)

plt.show()
```

```python
y_beijing = [random.uniform(8,14) for _ in range(60)]
plt.plot(x[::5], y_beijing[::5])

plt.show()
```

```python
plt.figure(figsize=(20,10), dpi=100)
plt.plot(x, y)
plt.plot(x, y_beijing)
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5], x_label[::5])
plt.yticks(range(40)[::5])
plt.title("xixixi")
plt.xlabel("one hour")
plt.ylabel("℃")
plt.show()
```
